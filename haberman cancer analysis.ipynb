univariate analysis

```python
sns.FacetGrid(cancer,hue="Survival",size = 5)\
.map(sns.distplot,"Age")\
.add_legend();
mlt.show()
```

```python
sns.FacetGrid(cancer,hue="Survival",size = 10)\
.map(sns.distplot,"Nodes")\
.add_legend();
mlt.show()
```

```python
sns.FacetGrid(cancer,hue="Survival",size = 5)\
.map(sns.distplot,"Year")\
.add_legend();
mlt.show()
```

```python
sns.boxplot(x='Survival',y='Age',data= cancer)
mlt.show()
sns.boxplot(x='Survival',y='Year',data=cancer)
mlt.show()
sns.boxplot(x='Survival',y='Nodes',data=cancer)
mlt.show()
```

```python
sns.violinplot(x= 'Survival',y= 'Age',data = cancer,height = 10)
mlt.show()
sns.violinplot(x='Survival',y='Year',data = cancer,height = 10)
mlt.show()
sns.violinplot(x='Survival',y='Nodes',data = cancer,height = 10)
mlt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(cancer, hue = 'Survival' , height = 6)\
 .map(mlt.scatter,'Age','Year')\
 .add_legend()
mlt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(cancer, hue='Survival', height = 5)
mlt.show()
```

```python
sns.jointplot(x ='Year', y = 'Age' ,data = cancer,kind = 'kde')
mlt.show()
```
